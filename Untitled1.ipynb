```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
national = pd.read_csv("indian-national-level-election.csv")
state = pd.read_csv("indian-state-level-election.csv")
```

```python
national.head()
state.head()
```

```python
national.columns
state.columns
```

```python
national.info()
state.info()
```

```python
national.describe()
state.describe()
```

```python
national.drop_duplicates(inplace=True)
state.drop_duplicates(inplace=True)
```

```python
national.duplicated().sum()
state.duplicated().sum()
```

```python
national['cand_sex'] = national['cand_sex'].fillna('Unknown')
state['cand_sex'] = state['cand_sex'].fillna('Unknown')
```

```python
national['cand_sex'].isnull().sum()
state['cand_sex'].isnull().sum()
```

```python
national['cand_sex'].unique()
state['cand_sex'].unique()
```

```python
national['turnout'] = national['totvotpoll'] / national['electors']
state['turnout'] = state['totvotpoll'] / state['electors']
```

```python
national.head()
national['turnout'].head()
```

```python
national['turnout'].describe()
state['turnout'].describe()
```

```python
national['turnout'].max()
state['turnout'].max()
```

```python
plt.figure()
national.groupby('year')['totvotpoll'].sum().plot(label='National')
state.groupby('year')['totvotpoll'].sum().plot(label='State')
plt.legend()
plt.title("Voting Trend")
plt.show()
```

```python
national.groupby('st_name')['totvotpoll'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top States - National")
plt.show()
```

```python
state.groupby('st_name')['totvotpoll'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top States - State")
plt.show()
```

```python
national['partyname'].value_counts().head(10).plot(kind='bar')
plt.title("Top Parties - National")
plt.show()
```

```python
state['partyname'].value_counts().head(10).plot(kind='bar')
plt.title("Top Parties - State")
plt.show()
```

```python
national['cand_sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender - National")
plt.ylabel('')
plt.show()
```

```python
state['cand_sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender - State")
plt.ylabel('')
plt.show()
```

```python
sns.histplot(national['turnout'], kde=True, label='National')
sns.histplot(state['turnout'], kde=True, label='State')
plt.legend()
plt.title("Turnout Comparison")
plt.show()
```

```python
sns.heatmap(national.corr(numeric_only=True), annot=True)
plt.title("Correlation - National")
plt.show()
```

```python
sns.heatmap(state.corr(numeric_only=True), annot=True)
plt.title("Correlation - State")
plt.show()
```

```python
# auto-detect column
if 'pc_name' in state.columns:
    group_cols = ['year', 'pc_name']
elif 'constituency' in state.columns:
    group_cols = ['year', 'constituency']
elif 'ac_name' in state.columns:
    group_cols = ['year', 'ac_name']
else:
    group_cols = ['year']

state_winners = (
    state.sort_values('totvotpoll', ascending=False)
         .groupby(group_cols)
         .first()
)

state_winners.head()
```
